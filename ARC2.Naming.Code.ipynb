```python
#Load the appropriate modules 
import numpy as np
import pandas as pd
```

```python
#Define range of latitudes, longitudes and pixelsize
startlat = -40
startlong = -20
endlat = 40
endlong = 55
pixsize = 0.1
```

```python
#Define number of latitude and longitude steps
step_lat = len(np.arange(startlat, endlat + pixsize, pixsize))
step_lon = len(np.arange(startlong, endlong + pixsize, pixsize))
print(step_lat)
print(step_lon)
```

```python
#Latitudes
lat = np.linspace(startlat, endlat, step_lat)
print(lat)
```

```python
#Longitudes
lon = np.linspace(startlong, endlong, step_lon)
print(lon)
```

```python
#Latitudes
lats = np.arange(startlat, endlat + pixsize, pixsize)
lats = np.around(lats, 1)

#Longitudes
longs = np.arange(startlong, endlong + pixsize, pixsize)
longs = np.around(longs, 1)

#Latitude column
latitude = np.repeat(lats, len(longs))

#Longitude column
longitude = np.tile(longs, len(lats))
```

```python
#create the pixelid which is a combination of latitude and longitude
pixel_id = [None] * len(latitude)

for i in range(len(latitude)):
    pixel_id[i] = str(latitude[i]) + "," + str(longitude[i])
```

```python
#Create the grid, a dataframe comprising of pixelid, latitude and longitude
grid = pd.DataFrame(list(zip(pixel_id, latitude, longitude)), 
                   index = pixel_id,                               # row names
                   columns = ["PixelID", "Latitude", "Longitude"]  # column names
                   )
#Print a few rows
grid.head()
```

```python
#Adding zeroes to the latitude and longitude
latitude=[]
for i in range(1,100):
    if i<10:
        i='00'+str(i)
        latitude.append(i)
    else:
        i='0'+str(i)
        latitude.append(i)
        

for i in range(100,802):
    i=str(i)
    latitude.append(i)
    

longitude=[]
for i in range(1,100):
    if i<10:
        i='00'+str(i)
        longitude.append(i)
    else:
        i='0'+str(i)
        longitude.append(i)
        

for i in range(100,752):
    i=str(i)
    longitude.append(i)
```

```python
grid_lats = np.repeat(latitude, len(longitude))
grid_longs = np.tile(longitude, len(latitude))

#Combining to obtain the unique names for the pixels
pixel_names = np.empty((len(grid_lats), 1), object)
for i in range(len(grid_lats)):
    pixel_names[i] = "Pixel" + grid_lats[i] + grid_longs[i]

#Add the pixelnames column to the grid
grid["PixelNames"] = pixel_names
#Print few rows
grid.head()
```

```python
#Saving the grid
    
grid.to_pickle("the_grid.pkl")
grid.to_csv('grid.csv')
```
